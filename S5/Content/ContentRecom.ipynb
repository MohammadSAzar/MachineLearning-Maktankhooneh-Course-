### Separation the Year

```python
titles = []
years = []

for item in movies_df['title']:
    title = item[:-7]
    year = item[len(item)-6:]
    year = year[1:5]
    titles.append(title)
    years.append(year)
    
movies_df['title'] = titles
movies_df['year'] = years
```

```python
movies_df.head()
# movies_df.shape
```

```python
genre_list = []
genre_columns = []
```

```python
for i in range(len(movies_df)):
    for genre in movies_df['genres'][i].split('|'):
        if genre not in genre_columns:
            genre_columns.append(genre)

for genre in genre_columns:
    movies_df[genre] = 0

print(genre_columns)
movies_df.head()
```

```python
for i in range(len(movies_df)):
    for genre in movies_df['genres'][i].split('|'):
        movies_df[genre][i] = 1
                
movies_df.head(10)
```

```python
ratings_df = ratings_df.drop(['timestamp'], axis=1)

ratings_df.head()
```

# Modelling

### Input

```python
user_input = [
            {'title':'Breakfast Club', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
user_input_df = pd.DataFrame(user_input)
user_input_df
```

```python
user_input_id = movies_df[movies_df['title'].isin(user_input_df['title'].tolist())]
user_input_id
```

```python
user_input_df = pd.merge(user_input_id, user_input_df)

user_input_df
```

```python
user_input_df = user_input_df.drop(['genres'], axis=1).drop(['year'], axis=1)

user_input_df
```

```python
user_input_df.reset_index()

user_input_df = user_input_df.drop(['movieId', 'title'], axis=1)

user_input_df
```

```python
user_genre_df = user_input_df.drop(['rating'], axis=1)
user_genre_df
```

```python
user_rating_df = user_input_df['rating']
user_rating_df
```

```python
user_profile = user_genre_df.transpose().dot(user_rating_df)
user_profile
```

```python
genres = movies_df.drop(['movieId', 'title', 'genres', 'year'], axis=1)

genres
```

```python
genres = genres.set_index(movies_df['movieId'])

genres
```

```python
recom = ((genres*user_profile).sum(axis=1))/(user_profile.sum())

recom
```

```python
recom = recom.sort_values(ascending=False)

recom.head()
```

```python
final_recom = movies_df.loc[movies_df['movieId'].isin(recom.head(20).keys())]

final_recom
```
