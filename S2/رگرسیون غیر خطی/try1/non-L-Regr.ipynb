# Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
```

# Read the data

```python
df = pd.read_csv('china_gdp.csv')
cdf = df[['Year', 'Value']]
xdata = cdf.Year
ydata = cdf.Value
plt.scatter(xdata, ydata, color='red')
```

# Introduce the model

```python
def sigmoidal_model(x, a, b, c):
    y = c/(1+np.exp(a*(x-b)))
    return y

yexample = sigmoidal_model(xdata, -0.2, 2500, 1000)
plt.scatter(xdata, yexample, color='blue')
```

# Normalize the data

```python
xdata_norm = xdata/max(xdata)
ydata_norm = ydata/max(ydata)
plt.scatter(xdata_norm, ydata_norm, color='purple')
```

# Fitting the model

```python
from scipy.optimize import curve_fit

popt, pcov = curve_fit(sigmoidal_model, xdata_norm, ydata_norm)
print("a: ", popt[0])
print("b: ", popt[1])
print("c: ", popt[2])
```

# Draws

```python
ymodel = sigmoidal_model(xdata_norm, *popt)

plt.scatter(xdata_norm, ydata_norm, color='black')
plt.scatter(xdata_norm, ymodel, color='orange')
```

# Accuracy & Error

```python
overparameterize_number = np.linalg.cond(pcov)
fitting_standard_error = np.sqrt(np.diag(pcov))

print(overparameterize_number, fitting_standard_error)
```
