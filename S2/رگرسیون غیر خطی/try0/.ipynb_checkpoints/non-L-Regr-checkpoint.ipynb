# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Samples of non-Linear

```python
x = np.arange(-5, 5, 0.2)
y = 5*np.power(2, x) + 2*x + 5
ynoise = 4*np.random.normal(size=x.size)
ydata = y + ynoise

plt.plot(x, y, 'r')
plt.plot(x, ydata, 'bo')
```

```python
x = np.arange(-5, 5, 0.2)
y = np.exp(0.5*x)
ynoise = 0.2*np.random.normal(size=x.size)
ydata = y + ynoise

plt.plot(x, y, 'r')
plt.plot(x, ydata, 'bo')
```

```python
#### x = np.arange(-5, 5, 0.2)
y = 2*x - 1/(2*np.power(3.1, x) + np.log(x+13))
ynoise = 1.1*np.random.normal(size=x.size)
ydata = y + ynoise

plt.plot(x, y, 'r')
plt.plot(x, ydata, 'bo')
```
