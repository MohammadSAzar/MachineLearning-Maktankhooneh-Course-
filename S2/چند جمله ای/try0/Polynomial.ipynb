## choose the cdf + choose the parameters

```python
cdf = df[['ENGINESIZE', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
viz = cdf[['ENGINESIZE', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
viz.hist()
plt.subplots_adjust(left=0.1,
                    bottom=0.9, 
                    right=2, 
                    top=2.2, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

## 80 - 20 data setting

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]

print(train)
print(test)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='red')
ax1.scatter(test.ENGINESIZE, test.CO2EMISSIONS, color='green')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emission')
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(train.FUELCONSUMPTION_COMB_MPG, train.CO2EMISSIONS, color='orange')
ax1.scatter(test.FUELCONSUMPTION_COMB_MPG, test.CO2EMISSIONS, color='blue')
plt.xlabel('Combined Fuel Cons MPG')
plt.ylabel('CO2 Emission')
plt.show()
```

## Polynomial Regression >>>> Data Preparation

```python
from sklearn.preprocessing import PolynomialFeatures

train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])

poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)

train_x_poly
```

```python
from sklearn import linear_model

clf = linear_model.LinearRegression()
clf.fit(train_x_poly, train_y)

print('Coefficients: ', clf.coef_)
print('Intercept: ', clf.intercept_)
```

## Polynomial Regression >>>> Diagram

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='red')

xx = np.arange(1, 8, 0.1)
yy = clf.intercept_[0] + clf.coef_[0][1]*xx + clf.coef_[0][2]*np.power(xx, 2)
plt.plot(xx, yy, color='green')

plt.xlabel("Engine Size")
plt.ylabel("CO2 Emission")
```

# Evaluation (Prediction)

```python
test_x_poly = poly.fit_transform(test_x)
test_y_ = clf.predict(test_x_poly)


print("MAE: ", np.mean(test_y_ - test_y))
print("Residual Sum of Squares: ", np.mean((test_y_ - test_y)**2))
print("Variance Score:", clf.score(test_x_poly, test_y))
```
