# Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
```

# Read the data

```python
df = pd.read_csv('house_price.csv')
df.head(10)
# df.shape[0]
```

# Clean the data

```python
### Null >>> empty string
df = df.fillna('')

### Remove data without address
df = df[df['Address'] != '']

### Drop duplicates
df = df.drop_duplicates()


### Remove data with unrealistic area
df['Area'] = df['Area'].str.replace(',', '')
for x in df.index:
    df.loc[x, 'Area'] = int(df.loc[x, 'Area'])

realistic_limit = 1000

for x in df.index:
    if df.loc[x, 'Area'] > realistic_limit:
        df.drop(x, inplace=True)

### Remove data with unrealistic area
df.reset_index(drop=True, inplace=True)

df.sort_values(by='Area', ascending=False).head(10)
df
# df.shape[0]
```

# Encoding the data

###  Assign numbers to booleans (elevator - warehouse - parking)

```python
df_encoded_booleans = df

descriptors = ['Parking', 'Warehouse', 'Elevator']
for descriptor in descriptors:
    df_encoded_booleans[descriptor] = df_encoded_booleans[descriptor].astype(int)

df_encoded_booleans
```

###  One-Hot encoding for non-boolean categoricals (Address)

```python
address_dummies = pd.get_dummies(df_encoded_booleans['Address'], prefix='Address')
df_encoded_booleans_and_non_booleans = pd.concat([df_encoded_booleans, address_dummies], axis=1)
df_encoded_booleans_and_non_booleans = df_encoded_booleans_and_non_booleans.drop('Address', axis=1)

df_encoded_booleans_and_non_booleans
```

```python
column_list = df_encoded_booleans_and_non_booleans.columns.tolist()

for descriptor in column_list[7:]:
    df_encoded_booleans_and_non_booleans[descriptor] = df_encoded_booleans_and_non_booleans[descriptor].astype(int)

df_encoded_booleans_and_non_booleans
```

# Feature selecting

### Correlation analysis

```python
correlations = df_encoded_booleans_and_non_booleans.corr()['Price'].sort_values(ascending=False)

cdf = df_encoded_booleans_and_non_booleans[['Area', 'Room', 'Price', 'Price(USD)']]

plt.scatter(cdf.Area, cdf.Price, color='black')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

plt.scatter(cdf.Room, cdf.Price, color='red')
plt.xlabel('Room')
plt.ylabel('Price')
plt.show()


#correlations.head(15)
```

# 80-20

```python
msk = np.random.rand(len(cdf)) < 0.8
train = cdf[msk]
test = cdf[~msk]
print(test, train)
```

# Regression

## Linear

#### Train

```python
from sklearn import linear_model

x_linear_train = np.asanyarray(train[['Area', 'Room']])
y_linear_train = np.asanyarray(train[['Price']])

regr = linear_model.LinearRegression()
regr.fit(x_linear_train, y_linear_train)

print('coefficients: ', regr.coef_)
print('intercept: ', regr.intercept_)
print('variance score: ', regr.score(x_linear_train, y_linear_train))
```

#### Test

```python
x_linear_test = np.asanyarray(test[['Area', 'Room']])
y_linear_test = np.asanyarray(test[['Price']])
y_linear_test_predicted = regr.predict(x_linear_test)

print('residual sum of squares: ', np.mean((y_linear_test_predicted-y_linear_test)**2))
print('variance score: ', regr.score(x_linear_test, y_linear_test))
```

## Polynomial

#### Train

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

x_poly_train = np.asanyarray(train[['Area', 'Room']])
y_poly_train = np.asanyarray(train[['Price']])

# making polynomial matrix
poly = PolynomialFeatures(degree=2)
x_poly_train_transform = poly.fit_transform(x_poly_train)

clf = linear_model.LinearRegression()
clf.fit(x_poly_train_transform, y_poly_train)

print('coefficients: ', clf.coef_)
print('intercept: ', clf.intercept_)
print('variance score: ', clf.score(x_poly_train_transform, y_poly_train))
```

#### Test

```python
x_poly_test = np.asanyarray(test[['Area', 'Room']])
y_poly_test = np.asanyarray(test[['Price']])

# making polynomial matrix
x_poly_test_transform = poly.fit_transform(x_poly_test)

y_poly_test_predicted = clf.predict(x_poly_test_transform)

print('MAE: ', np.mean(y_poly_test_predicted-y_poly_test))
print('residual sum of squares: ', np.mean((y_poly_test_predicted-y_poly_test)**2))
print('variance score: ', clf.score(x_poly_test_transform, y_poly_test))
```
